```python
import numpy as np
import pandas as pd

with open("LLCP2016.ASC ", "r") as f:
    for i in range(5):
        print(f.readline())
```

```python
colspecs = [
    (0, 2),     # _STATE
    (18, 26),    #DATE
    (1977, 1979), # _AGE80
    (119, 120),   # SEXVAR
    (1970, 1971),   # _RACE
    (92, 94),   # MENTHLTH
    (199, 200),   # ECIGNOW2    
    (197, 198),   # USENOW3
    (2001, 2002),   # _RFSMOK3
    (2008, 2009),   # _RFBING6
    (389, 391),    # MARIJAN1
    (1746, 1756)    #LLCPWT
]

colnames = [
    "_STATE", "DATE", "_AGE80", "SEXVAR", "_RACE", "MENTHLTH", "ECIGNOW2", "USENOW3", "_RFSMOK3",
    "_RFBING6", "MARIJAN1", "LLCPWT"
]
df = pd.read_fwf("LLCP2016.ASC ", colspecs=colspecs, names=colnames)

# Preview
print(df.head())
```

```python
sexvar_counts = df['SEXVAR'].value_counts()
print(sexvar_counts)
```

```python
df['ECIGNOW2'] = df['ECIGNOW2'].replace({1:2, 2:3, 3:4, 4:1}) #2021,2020,2018,2017,2016  ECIGNOW1
```

```python
df = df[df['ECIGNOW2'].notna()]
df = df[df["SEXVAR"] != 7]
df = df[df["SEXVAR"] != 9]
df.to_csv("BRFSS2016_cleaned.csv", index=False)
```
